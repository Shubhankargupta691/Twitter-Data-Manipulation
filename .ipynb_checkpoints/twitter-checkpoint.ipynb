```python
data = pd.read_csv('C:\\Users\SHUBHANKAR GUPTA\\Documents\\DATASCIENCE\\DATA SETS\\twitter\\Survey Monkey Output.csv')
data
```

Making a copy of data

```python
data_modified = data.copy()
data_modified
```

Droping Unwanted Columns

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address',
       'First Name', 'Last Name', 'Custom Data 1']
columns_to_drop
```

Droping the unwanted columns from our data-set

```python
data_modified  = data_modified.drop(columns= columns_to_drop)
data_modified
```

```python
data_modified.columns
```

Making first eight columns as "Identifieres"
And Others as "Variables"

```python
id_vars = list(data_modified.columns[:8])
value_vars = list(data_modified.columns[8: ])
```

```python
data_melted = data_modified.melt(id_vars= id_vars, value_vars= value_vars, var_name="Qusetions + SubQuestions", value_name="Answer")
data_melted
```

```python
questions_import= pd.read_csv('C:\\Users\SHUBHANKAR GUPTA\\Documents\\DATASCIENCE\\DATA SETS\\twitter\\Questions.csv')
questions_import
```

Here  "inplace = True "  means dispaly questions that I have modified

```python
questions = questions_import.copy()
questions.drop(columns=["Raw Question", "Raw Subquestion", "Subquestion"], inplace=True)
questions
```

```python
dataset_merged = pd.merge(left=data_melted, right=questions, how="left", left_on="Question + Subquestion", right_on="Question + Subquestion")
print("Original Data", len(data_melted))
print("Merged Data", len(dataset_merged))
dataset_merged
```

```python
respondents = data_merged[data_merged["Answer"].notna()]
respondents = respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
respondents.rename(columns={"Respondent ID":"Respondents"}, inplace=True)
respondents
```
